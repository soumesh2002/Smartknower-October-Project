### Handling missing values

```python
missing_values = data.select_dtypes(include=['object']).columns
```

```python
missing_values
```

```python
for x in missing_values:
    percentage = len(data[data[x] == '?']) / data.shape[0] * 100
    print(x, f"{percentage:.2f}%")
```
